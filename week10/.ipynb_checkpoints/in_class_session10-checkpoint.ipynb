```python
from collections import Counter
text = "april is the cruelest month breeding lilacs out of the dead"
words = text.split()
count = Counter()
for item in words:
    sounds_list = pr.phones_for_word(item)
    if len(sounds_list) > 0:
        #print("\n".join(sounds_list[0].split()))
        count.update(sounds_list[0].split())
count.most_common()
```
